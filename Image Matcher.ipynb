```python
import cv2
import os
import numpy as np
from skimage.metrics import structural_similarity as ssim
from google.colab.patches import cv2_imshow

def compare_images(image1, image2):
    # Load images in grayscale
    img1 = cv2.imread(image1, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(image2, cv2.IMREAD_GRAYSCALE)

    # Check if both images were loaded successfully
    if img1 is None or img2 is None:
        print(f"Failed to load images: {image1}, {image2}")
        return 0.0

    # Resize images to the same dimensions (256x256)
    img1 = cv2.resize(img1, (256, 256))
    img2 = cv2.resize(img2, (256, 256))

    # Compute the Structural Similarity Index (SSIM)
    ssim_score = ssim(img1, img2)

    return ssim_score

def find_similar_images(folder1, folder2):
    for folder1_img in os.listdir(folder1):
        folder1_img_path = os.path.join(folder1, folder1_img)
        max_ssim_scores = []  # Store top 3 SSIM scores
        matching_img_paths = []  # Store corresponding matching image paths

        for folder2_img in os.listdir(folder2):
            folder2_img_path = os.path.join(folder2, folder2_img)
            ssim_score = compare_images(folder1_img_path, folder2_img_path)

            # Find the position to insert the current SSIM score
            insert_position = len(max_ssim_scores)
            for i, score in enumerate(max_ssim_scores):
                if ssim_score > score:
                    insert_position = i
                    break

            # Insert the current SSIM score and corresponding image path
            max_ssim_scores.insert(insert_position, ssim_score)
            matching_img_paths.insert(insert_position, folder2_img_path)

            # Keep only the top 3 scores and image paths
            max_ssim_scores = max_ssim_scores[:3]
            matching_img_paths = matching_img_paths[:3]

        original_img = cv2.imread(folder1_img_path)

        # Display the title and image information before showing the images
        print("Original Image:")
        print(f"Image Path: {folder1_img_path}")
        print(f"Top 3 SSIM Scores: {[round(score, 4) for score in max_ssim_scores]}")

        for i, (ssim_score, matching_img_path) in enumerate(zip(max_ssim_scores, matching_img_paths)):
            if ssim_score >= 0.4:  # If SSIM Score is greater than or equal to 50%
                print(f"Matching Image {i+1}:")
                print(f"Image Path: {matching_img_path}")
                print("--------------------------------------------------")

                # Load matching image and resize both images to 256x256 for display
                matching_img = cv2.imread(matching_img_path)
                original_img_resized = cv2.resize(original_img, (256, 256))
                matching_img_resized = cv2.resize(matching_img, (256, 256))

                # Display original and matching images
                cv2_imshow(original_img_resized)
                cv2_imshow(matching_img_resized)
            else:
                print(f"Matching Image {i+1}: Image not found in the Database")
                print("--------------------------------------------------")

        # Wait for the user to press any key to continue
        cv2.waitKey(0)
        cv2.destroyAllWindows()

if __name__ == "__main__":
    folder1_path = "/content/drive/MyDrive/Image_Matcher/Sinp_finder_Images/Test"
    folder2_path = "/content/drive/MyDrive/Image_Matcher/Sinp_finder_Images/Database"

    find_similar_images(folder1_path, folder2_path)
```
